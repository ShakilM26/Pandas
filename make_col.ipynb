```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'names':['steve', 'stephen', 'smith', 'lucy', 'rimix', 'silvia'],
                   'grade':['A+', 'A+', 'A', 'A', 'A+', 'A'],
                   'salary': [40000, 42000, 45000, 35000, 30000, 33000],
                   'country': ['usa', 'usa', 'uk', 'russia', 'russia', 'usa']})
```

```python
# make new col using pandas 'where' function
# Condition: The values that fit the condition remain the same 

df['class'] = df.salary.where(df.salary < 40000, 'Level_1')
df
```

```python
# numpy 'where' function

df['salary_range'] = np.where((df.salary >= 40000) & (df.grade=='A'), 'High', 'Medium')
df
```

```python
# numpy 'select' function

condi = [(df.salary >= 40000) & (df.country == 'usa')]
values = ['Rich']

df['new'] = np.select(condi, values, default='Medium')
df
```
