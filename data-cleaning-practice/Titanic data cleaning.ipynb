```python
import pandas as pd
```

```python
df = pd.read_csv('full.csv')
df.info()
```

```python
cols = ['PassengerId', 'SibSp','Parch','WikiId','Name_wiki','Age_wiki']

df.drop(cols, inplace=True, axis=1)
```

```python
df.info(memory_usage='deep')
```

```python
import seaborn as sns
sns.heatmap(df.isna())

# if some case column names aren't easily readable
# sns.heatmap(df.isna().transpose)
```

```python
# how many percentages  values are missing from individual 
# columns. This can be useful while handling these missing values.

import numpy as np
print('Amount of missing values in -:')

for col in df.columns:
    percentage_missing = np.mean(df[col].isna())
    print(f'{col}: {round(percentage_missing*100)}%')
```

There is no standard way for dealing with the missing data. The only way is to look at the individual column, amount of missing values in it and importance of that column in the future.
