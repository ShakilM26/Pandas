```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('weather.csv')
df.set_index('Day', inplace=True)
```

```python
# change the data type

df['temp']=pd.to_numeric(df['temp'], errors='coerce')
df['w speed']=pd.to_numeric(df['w speed'], errors='coerce')
df
```

```python
# Replace the negative value

df['temp']=df['temp'].replace([-40], 31)
df 
```

```python
df['temp'].mean()
```

```python
# Interpolation have a huge power when we use it handling missing values

df.interpolate(inplace=True)
df
```

```python
# We can also change the nan value with 'No Event'

df.replace({'event': np.nan}, 'No Event', inplace=True)
df
```

```python
# We can change all No Event values into a same value

df.replace({'No Event': 'Sunny'}, inplace=True)
df
```

```python
# Still have a wrong value

df.replace({'Nice': 'Cloudy'}, inplace=True)
df
```

```python
# Finally we should change the data type of our numeric columns 

df['temp']=df['temp'].astype('int64')
df['w speed']=df['w speed'].astype('int64')
df
```
