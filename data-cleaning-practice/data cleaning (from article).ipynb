```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_to':['LoNDon_paris','MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis','Brussels_LondOn'],
                  'FlightNumber':[10045.0,np.nan,10065.0,np.nan,10085.0],
                  'RecentDelays':[[23,47],[],[24,43,87],[13],[67,32]],
                  'Airline':['KLM(!)','(12)','(British Airways. )','12. Air France','"Swiss Air"']})
df
```

```python
df1 = df['From_to'].str.split('_', expand=True)

# Here expand split string into multiple columns
df1
```

```python
# rename column 

df1 = df1.rename(columns={0: 'Departure', 1:'Arrival'})
df1
```

```python
# although the names of 'Departure' and 'Arrival' cols values aren't proper format

for i in df1.columns:
    df1[i] = df1[i].str.capitalize()
df1
```

```python
df = df.drop('From_to', axis=1)
df = pd.concat([df,df1], axis=1)
df
```

```python
# Flight Number col
# we can increse the numbers compare to before value

for i in range(len(df['FlightNumber'])):
    if pd.isnull(df.loc[i, 'FlightNumber'])==True:
        df.loc[i, 'FlightNumber']=df.loc[i-1, 'FlightNumber']+10
df
```

```python
df2 = pd.DataFrame(df['RecentDelays'].tolist())
# using tolist(), it's break into different parts

df2 = df2.rename(columns={0:'Delay1',1:'Delay2',2:'Delay3'})
df2
```

```python
# filling nan values with it's previous rows and cols values

df2 = df2.ffill(axis=0)
df2 = df2.ffill(axis=1)
df2
```

```python
# now add into the main dataframe

df = df.drop('RecentDelays', axis=1)
```

```python
df = pd.concat([df, df2], axis=1)
df
```

```python
# Airline

import re

def clean(x):
    x = re.sub(r'[0-9]', '', x)
    x = x.strip()
df['Airline'] = df['Airline'].apply(clean)
```
