```python
import pandas as pd
```

```python
df = pd.read_csv('items.csv')
```

```python
df.columns
```

```python
# In description col our target word is 'used car'

df[df['description'].str.contains('used car')]
```

```python
# also we can search like 

df[df['description'].str.contains('used') & df['description'].str.contains('car')]
```

```python
# using apply & len function
 
df[df['description'].apply(lambda x: len(x) > 12)]
```

```python
# make proper numeric and remove unwanted sign

df[df['price'].apply(lambda x: x.isnumeric()==True)]
```

```python
df[df['description'].str.count('used') < 1]
```

```python
id = pd.read_csv('identity.csv')
```

```python
# Logical operators

id[id.cgpa < 3]
```

```python
# Logical operators also works on string.
# Only comes after 'Shakil' in alphabetical order

id[id.Name > 'Shakil']
```

```python
# Multiple Condition

id[(id.cgpa < 3.00) & (id.code > 10)]
```

```python
# Multiple Condition but using tilde and apply function

id[~(id.country == 'bangladeshi') | (id.country.apply(lambda x: len(x) < 6))]
```

```python
# isin

names = ['Shakil', 'Fenellas', 'Stephen']

id[id.Name.isin(names)]
```

```python
# string accesor
# startswith & endswith

id[id.Name.str.endswith('a')]
```

```python
# Using tilde ~ 

id[~id.country.str.startswith('b')]
```

```python
# string accesor

id[id.country.str.contains('usa')] 
```

```python
# Query 

id.query('cgpa < 3 & country == "winterfell"')
```

```python
# nlargest, nsmallest. both are same syntax

id.nlargest(2, 'cgpa')
```
