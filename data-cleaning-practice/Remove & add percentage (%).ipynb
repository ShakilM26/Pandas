#### Now remove % and change cols data type

```python
df['Yearly_Change'] = list(map(lambda x: x[:-1], df['Yearly_Change'].values))
df['World_Share'] = list(map(lambda x: x[:-1], df['World_Share'].values))
```

```python
df['Yearly_Change'] = [float(x) for x in df['Yearly_Change'].values]
df['World_Share'] = [float(x) for x in df['World_Share'].values]
```

```python
df.info()
```

#### How to add percentage 

```python
df['Yearly_Change'] =  df['Yearly_Change'].map("{:,.2f}%".format)
df.head()
```
