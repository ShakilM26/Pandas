```python
import pandas as pd
```

```python
df = pd.read_csv('apple.csv', parse_dates=['Date'], index_col='Date')
df.head(5)
```

```python
retrieve = (df['Date'] >= '2022-10-01') & (df['Date'] <= '2022-10-12')
df.loc[retrieve]
```

```python
# day difference 

df.index[0] - df.index[-1]
```

```python
# retrive september's 27

df.loc['2022-09-27']
```

```python
df.loc['2021-10'].Close.mean()
```

```python
# for close col every month's average value 

df.Close.resample('M').mean().head(5)
```

```python
df.Close.resample('M').mean().plot()
```

```python
df.Close.resample('Q').mean().plot(kind='bar')
```

```python
df.Close.plot()
```

```python
import seaborn as sns 

sns.set(rc={'figure.figsize':(11,6)})
df['Open'].plot(linewidth=4)
```

```python
cols=['Open', 'High', 'Low']
ax=df[cols].plot(marker='D', alpha=1, linestyle='solid', figsize=(11,9), subplots=True)

for a in ax:
  a.set_ylabel('Daily Graph')
  a.set_xlabel('Date & Month')
```
