* Data wrangling involves processing the data in various formats like - merging,grouping,concatenating etc. 
* For the purpose of analysing or getting them ready to be used with another set of data.

### Data Merging

```python
import pandas as pd

left = pd.DataFrame({'id':[101,201,301,401,501],
                    'name':['peter','tony','clark','steve','stephen'],
                    'occup':['student','engineer','journalist','captain','doctor']})
right = pd.DataFrame({'id':[101,201,301,401,501],
                     'name':['miles','wanda','eddie','bruce','mark'],
                     'occup':['student','teacher','journalist','ceo','scientist']})

merging_on_occup = pd.merge(left, right, how='inner', on='occup')
merging_on_occup
```

```python
merging_on_id = pd.merge(left, right, how='inner', on='id')
merging_on_id
```

### Concatenating

```python
concat = pd.concat([left, right])
concat
```

### Grouping

```python
score = {'name':['mohosin','kawser','zakir','mahfuj','shakil','mintu','rafsan','mushfiq'],
        'sub':['Algorithm','Algorithm','Algorithm','OS','Data Mining','Data Mining','System Design','System Design'],
        'marks':[92,93,90,85,88,92,84,86]}
df = pd.DataFrame(score)
df
```

```python
grp = df.groupby('sub')
grp.get_group('Algorithm')
```
