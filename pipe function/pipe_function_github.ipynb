```python
import pandas as pd
import numpy as np
```

```python
employees = pd.DataFrame({'emp_name':['Almirr Shakil', 'Belal Hossain', 'Kawser Mahbub', 'Mohosin Kabir', 'Mintu Hossain',
                                     'Sejuti Anjum', 'Samia Binte', 'Zakir Hossain', 'Farjana Tithi', 'Imran Hossain'],
                          
                          'job_title':['Data Analyst', 'Software Engineer', 'Lecturer', 'Lecturer', 'ML Engineer',
                                      'Software Engineer', 'Data Analyst', 'Chief Executive', 'ML Engineer', 'Web Developer'],
                          'company': ['Cramstack', 'Brain Station 23', 'BUBT','BUBT', 'Growuply','Data Soft','Micro1', 'STDL',
                                     'Data Soft', 'Mediusware'],
                          'from':['Rajshahi','Rajshahi','Dhaka','Dhaka','Jaypurhat','Dhaka','Dhaka','Pabna','Chandpur',
                                  'Satkhira'],
                          'joining_date': ['12/22/2022','04/23/2022', '08/10/2021','09/25/2022','12/12/2022','05/26/2022',
                                              '04/10/2023','02/02/2021', '11/10/2021', '10/31/2022'],
                          'From_office (min)': [60,65,20,15,50,35,60,80,10,15]})
employees
```

```python
# first func

def get_first_and_last_name(df, col_name):
    new_df = df.copy()
    
    split_the_name_col = new_df[col_name].str.split(' ', expand=True)
    
    new_df['First_name'] = split_the_name_col.get(0)
    new_df['Last_name'] = split_the_name_col.get(1)
    
    return new_df

after_split_col = get_first_and_last_name(employees, 'emp_name')
after_split_col
```

```python
employees['joining_date'] = pd.to_datetime(employees['joining_date'])
employees.info()
```

```python
# second func

def get_calendar_info(df, col_name):
    
    calendar_info = df[col_name]
    
    new_df = df.copy()
    
    new_df["Day"] = calendar_info.dt.day
    new_df["Month"] = calendar_info.dt.month
    new_df["Year"] = calendar_info.dt.year
    new_df["Day_of_week"] = calendar_info.dt.day_name()
    new_df["Month_name"] = calendar_info.dt.month_name()
    return new_df

calendar_info = get_calendar_info(employees, 'joining_date')
calendar_info
```

```python
def short_info_each_employee(info):
    
    # select my desire information
    name = info.emp_name
    job_title = info.job_title
    company = info.company
    duration = info["From_office (min)"]
    
    
    emp_info = f'''{name} is working as a {job_title} in 
    {company}. His office is {duration} minutes away 
    from his home.'''
    
    return emp_info
```

```python
# final_info

def employee_info(employees):
    
    final_df = employees.copy()
    
    final_df['Info'] = final_df.apply(lambda info: short_info_each_employee(info), axis=1)
    
    return final_df

final_info = employee_info(employees)
final_info
```

```python
preprocessing_all_func = (employees.pipe(get_first_and_last_name, 'emp_name').
                         pipe(get_calendar_info, 'joining_date').
                         pipe(employee_info))
preprocessing_all_func
```
